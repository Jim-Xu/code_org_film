```python
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
file_6am = xr.open_dataset('/data/keeling/a/xx24/d/proj_film/partmc/scenarios/1_urban_plume/out/urban_plume_00000007_process.nc', engine='netcdf4')
file_18pm = xr.open_dataset('/data/keeling/a/xx24/d/proj_film/partmc/scenarios/1_urban_plume/out/urban_plume_00000019_process.nc', engine='netcdf4')
```

```python
diam_edges = file_6am["diam_edges"][:]* 1e6
```

```python
bc_frac_edges = file_6am["bc_frac_edges"][:] * 100
diam_bc_dist_6am = file_6am["diam_bc_dist"][:] * 1e-6
diam_bc_dist_18pm = file_18pm["diam_bc_dist"][:] * 1e-6

fig, axs = plt.subplots(1, 2, figsize=(12, 5),sharex=True, sharey=True)

c1 = axs[0].pcolormesh(diam_edges, bc_frac_edges, diam_bc_dist_6am,
                       norm=colors.LogNorm(vmin=1e3, vmax=1e5), linewidths=0.1)
axs[0].set_xscale("log")
axs[0].set_xlabel(r"Dry Diameter $D_{\rm dry}$ / $\rm \mu m$")
axs[0].set_xlim(1e-2, 1e0)
axs[0].set_yscale("linear")
axs[0].set_ylabel(r"BC mass fraction $w_{\rm BC}$ / %")
axs[0].set_ylim(0, 80)
axs[0].grid(True)
axs[0].set_title("6 AM")

c2 = axs[1].pcolormesh(diam_edges, bc_frac_edges, diam_bc_dist_18pm,
                       norm=colors.LogNorm(vmin=1e3, vmax=1e5), linewidths=0.1)
axs[1].set_xscale("log")
axs[1].set_xlabel(r"Dry Diameter $D_{\rm dry}$ / $\rm \mu m$")
axs[1].set_xlim(1e-2, 1e0)
axs[1].set_yscale("linear")
axs[1].grid(True)
axs[1].set_title("18 PM")

fig.colorbar(c2, ax=axs, location='right', label=r"Number Conc. $n(D_{\rm dry},w_{\rm BC})$ / $\rm cm^{-3}$")
```
